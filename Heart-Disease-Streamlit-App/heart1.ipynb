```python
from scipy.stats import chi2_contingency
cat_cols=['Chest_Pain', 'Shortness_of_Breath', 'Fatigue', 'Palpitations','Dizziness', 'Swelling', 'Pain_Arms_Jaw_Back', 'Cold_Sweats_Nausea','High_BP', 'High_Cholesterol', 'Diabetes', 'Smoking', 'Obesity','Sedentary_Lifestyle', 'Family_History', 'Chronic_Stress', 'Gender','Age']
# Chi-Square test for each categorical column
for col in cat_cols:
    contingency_table = pd.crosstab(df[col], df['Heart_Risk'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    
    print(f"Column: {col}")
    print(f"Chi2: {chi2:.2f}, p-value: {p:.4f}")
    if p < 0.05:
        print("=> Dependent on target (significant)\n")
    else:
        print("=> Independent of target (not significant)\n")
```

```python
fig, axes = plt.subplots(4, 5, figsize=(15, 8))

for i, col in enumerate(cat_cols):
    sns.histplot(x=col, data=df, ax=axes[i//5, i%5])
    axes[i//5, i%5].set_title(col)

plt.tight_layout()
plt.show()
```

```python


from sklearn.model_selection import train_test_split
x=df.drop('Heart_Risk',axis=1)
y=df['Heart_Risk']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,f1_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
models={
    "logistic":LogisticRegression(),
    "knn":KNeighborsClassifier(n_neighbors=5),
    "GaussianNB":GaussianNB(),
    "DecisionTree":DecisionTreeClassifier(random_state=42),
    "svm":SVC(kernel='rbf')
}
```

```python
for name,model in models.items():
    model.fit(x_train,y_train)
    pred=model.predict(x_test)
    acc=accuracy_score(y_test,pred)
    cf=confusion_matrix(y_test,pred)
    f1=f1_score(y_test,pred)
    rp=classification_report(y_test,pred)
    print(f"{name},accuracy:{acc}confus:{cf}")
    
```
