```python

x=df.drop('label',axis=1).values
y=df['label'].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

```

```python
x_train=x_train.astype('float32')/255.0
x_test=x_test.astype('float32')/255.0
```

```python
xtranmg=x_train.reshape(-1,28,28)
xtestmg=x_test.reshape(-1,28,28)
```

```python
y_train_cat=to_categorical(y_train,10)
y_test_cat=to_categorical(y_test,10)
```

```python
perc=Sequential([
    Flatten(input_shape=(28,28)),
    Dense(10,activation='softmax')])
```

```python
perc.compile(optimizer='sgd',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history=perc.fit(xtranmg,y_train_cat,
                 epochs=5,batch_size=32,validation_data=(xtestmg,y_test_cat),verbose=1)
```

```python
loss,acc=perc.evaluate(xtestmg,y_test_cat,verbose=0)
print(acc)
```

```python
xtranmg=x_train.reshape(-1,28,28,1)
xtestmg=x_test.reshape(-1,28,28,1)
```

```python
cnn=Sequential([
    Conv2D(32,kernel_size=(3,3),activation='relu',input_shape=(28,28,1)),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(64,kernel_size=(3,3),activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Flatten(),
    Dense(128,activation='relu'),
    Dropout(0.5),
    Dense(10,activation='softmax')

])
```

```python
cnn.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history_cnn=cnn.fit(xtranmg,y_train_cat,
                 epochs=5,batch_size=32,validation_data=(xtestmg,y_test_cat),verbose=1)
```
